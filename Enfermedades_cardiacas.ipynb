```python
data.drop(["dataset", "fbs","restecg"],
          axis=1,
          inplace=True)
```

```python
data.columns
```

```python
for column in ['exang', 'slope', 'thal']: # Nos centraremos en las columnas categÃ³ricas identificadas anteriormente como que tienen valores faltantes: exang, slope, y thal don valores categoricos
    mode_value = data[column].mode()[0]
    data[column] = data[column].fillna(mode_value)
```

```python
for column in ['trestbps', 'chol', 'thalch', 'oldpeak', 'ca']: #Utilizaremos la mediana para cada columna para completar los NaN valores en trestbps, chol, thalch, oldpeak, y ca.
    median_value = data[column].median()
    data[column] = data[column].fillna(median_value)
```
